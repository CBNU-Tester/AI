```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('train_data_4.csv')

# Important가 1인 데이터를 모두 선택
df_important_1 = df[df['Important'] == 1]

# Important가 0인 데이터를 선택
df_important_0 = df[df['Important'] == 0]

# 각 비율에 맞게 Important가 0인 데이터 샘플링 함수 정의
def sample_data(ratio):
    target_count = len(df_important_1) * ratio
    sampled_data = pd.DataFrame()

    # 개수가 적은 Tag를 우선적으로 포함시키기 위해 Tag별로 샘플링
    tags = df_important_0['Tag'].value_counts().index.tolist()
    for tag in tags:
        tag_data = df_important_0[df_important_0['Tag'] == tag]
        if len(sampled_data) + len(tag_data) <= target_count:
            # 부족할 경우 전체 Tag 데이터를 포함
            sampled_data = pd.concat([sampled_data, tag_data])
        else:
            # 남은 샘플 수만큼만 샘플링
            remaining_count = target_count - len(sampled_data)
            sampled_data = pd.concat([sampled_data, tag_data.sample(remaining_count)])
            break

    return pd.concat([df_important_1, sampled_data])

# 1:1, 2:1, 3:1 비율로 데이터셋 준비
df_1 = sample_data(ratio=1)
df_2 = sample_data(ratio=2)
df_3 = sample_data(ratio=3)

# 준비된 데이터셋을 각각 CSV 파일로 저장
df_1.to_csv('train_data_4_1.csv', index=False, encoding='utf-8')
df_2.to_csv('train_data_4_2.csv', index=False, encoding='utf-8')
df_3.to_csv('train_data_4_3.csv', index=False, encoding='utf-8')

print("데이터셋이 1:1, 2:1, 3:1 비율로 각각 준비되고 저장되었습니다.")
```
